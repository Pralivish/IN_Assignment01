#### 2.Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.


```python
'''Python program to find numbers divisible by 7 and not multiples of 5 between 2000,3200'''

#Initializing the variables
a,b = 2000,3201

#for loop to find numbers divisible by 7 and not multiples of 5
for i in range(a,b):
    if (i%7==0) and (i%5!=0):
        print(i,end=',')
    
```

#### 4. Write a Python program to find the volume of a sphere with diameter 12 cm. Formula: V=4/3 *  * r 3


```python
'''Calculating the volume of the Sphere with diameter 12cm'''
#importing math package for using "pi"

from math import pi

d = 12                                      #diameter is denoted as "d"
r = d/2                              #representing radius as "r"

#volume of sphere v is:
x = 4/3                                     #x is 4/3
v = x*pi*r**3 

print(v)
```

#### 2. Create the below pattern using nested for loop in Python.

     *
     * *
     * * *
     * * * *
     * * * * *
     * * * *
     * * *
     * *
     *


```python
'''Python program to create patterns using nested for loop'''

for i in range(5):
    for j in range(i):
        print('*', end='')
    print('')

    
for i in range(5):
    for j in range(5-i):
        print('*', end='')
    print('')
     
```
